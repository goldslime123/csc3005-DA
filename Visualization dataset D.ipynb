```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
col_name = ['Profile factor','Symmetry','Crossflow',
            'Flow velocity P1','Flow velocity P2','Flow velocity P3','Flow velocity P4',
            'Speed of sound P1','Speed of sound P2','Speed of sound P3','Speed of sound P4',
             'Signal strength P1H','Signal strength P2H','Signal strength P3H','Signal strength P4H',
            'Signal strength P1T','Signal strength P2T','Signal strength P3T','Signal strength P4T',
            'Signal quality P1H','Signal quality P2H','Signal quality P3H','Signal quality P4H',
            'Signal quality P1T','Signal quality P2T','Signal quality P3T','Signal quality P4T',
            'Gain of P1H','Gain of P2H','Gain of P3H','Gain of P4H',
            'Gain of P1T','Gain of P2T','Gain of P3T','Gain of P4T',
            'Transit time of P1H','Transit time of P2H','Transit time of P3H','Transit time of P4H',
            'Transit time of P1T','Transit time of P2T','Transit time of P3T','Transit time of P4T',
            'HS']
```

```python
df = pd.read_csv('Meter D',
                 header=None,
                 sep='\t')
df.columns = col_name
df
```

# check if there is 0 in data

```python
# Count number of zeros in all columns of Dataframe
for column_name in df.columns:
    column = df[column_name]
    # Get the count of Zeros in column 
    count = (column == 0).sum()
    print(f'Count of zeros in column: {column_name} is : {count}')
```

# 2 Paths
1) remove rows

2) replace with median values

```python
df_remove = df.replace(0, np.nan)
df_remove = df_remove.dropna(how='any', axis=0) # axis = 0(row), = 1(columns)
df_remove = df_remove.replace(np.nan, 0)
df_remove
```

```python
df_median = df
for col_name in df.columns:
    df_median.loc[df[col_name] == 0, str(col_name)] = df[col_name].median()
df_median
```

# Boxplot for 1st column (df_median)

```python
val_1 = df_median['Profile factor']
plt.boxplot(val_1)
fig = plt.figure(figsize =(10, 7))
plt.show()

# finding the 1st quartile
q1 = np.quantile(val_1, 0.25)
 
# finding the 3rd quartile
q3 = np.quantile(val_1, 0.75)
med = np.median(val_1)
 
# finding the iqr region
iqr = q3-q1
 
# finding upper and lower whiskers
upper_bound = q3+(1.5*iqr)
lower_bound = q1-(1.5*iqr)
print('IQR:',iqr, 'Upper bound:',upper_bound, 'Lower bound:',lower_bound)

outliers = val_1[(val_1 <= lower_bound) | (val_1 >= upper_bound)]
print('The following are the outliers in the boxplot:{}'.format(outliers))
len(outliers)
```

```python
pca.explained_variance_ratio_
```

```python
explained_veriance = pca.explained_variance_ratio_
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.title('PCA components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Varience')
```

```python
import pandas as pd

# headers
flowVelocity_4Paths = [
    'Flow Velocity P1', 'Flow Velocity P2', 'Flow Velocity P3',
    'Flow Velocity P4'
]
speedOfSound_4Paths = [
    'Speed of Sound P1', 'Speed of Sound P2', 'Speed of Sound P3',
    'Speed of Sound P4'
]
signalStrength_4Paths = [
    'Signal Strength P1T', 'Signal Strength P1R', 'Signal Strength P2T',
    'Signal Strength P2R', 'Signal Strength P2T', 'Signal Strength P2R',
    'Signal Strength P2T', 'Signal Strength P2R'
]
turbulence_4Paths = [
    'Turbulence P1', 'Turbulence P2', 'Turbulence P3', 'Turbulence P4'
]
signalQuality_4Paths = [
    'Signal Quality P1T', 'Signal Quality P1R', 'Signal Quality P2T',
    'Signal Quality P2R', 'Signal Quality P3T', 'Signal Quality P3R',
    'Signal Quality P4T', 'Signal Quality P4R'
]
gain_4Paths = [
    'Gain P1T', 'Gain P1R', 'Gain P2T', 'Gain P2R', 'Gain P3T', 'Gain P3R',
    'Gain P4T', 'Gain P4R'
]
transitTime_4Paths = [
    'Transit Time P1T', 'Transit Time P2R', 'Transit Time P2T',
    'Transit Time P2R', 'Transit Time P3T', 'Transit Time P3R',
    'Transit Time P4T', 'Transit Time P4R'
]

flowVelocity_8Paths = [
    'Flow Velocity P1', 'Flow Velocity P2', 'Flow Velocity P3',
    'Flow Velocity P4', 'Flow Velocity P5', 'Flow Velocity P6',
    'Flow Velocity P7', 'Flow Velocity P8'
]
speedOfSound_8Paths = [
    'Speed of Sound P1', 'Speed of Sound P2', 'Speed of Sound P3',
    'Speed of Sound P4', 'Speed of Sound P5', 'Speed of Sound P6',
    'Speed of Sound P7', 'Speed of Sound P8'
]
gain_8Paths = [
    'Gain P1T', 'Gain P1R', 'Gain P2T', 'Gain P2R', 'Gain P3T', 'Gain P3R',
    'Gain P4T', 'Gain P4R', 'Gain P5T', 'Gain P5R', 'Gain P6T', 'Gain P6R',
    'Gain P7T', 'Gain P7R', 'Gain P8T', 'Gain P8R'
]

headerMeterA = ['Flatness Ratio', 'Symmetry', 'Crossflow'
                ] + flowVelocity_8Paths + speedOfSound_8Paths + [
                    'Average Speed of Sound'
                ] + gain_8Paths + ['Class']

headerMeterB = [
    'Profile Factor', 'Symmetry', 'Crossflow', 'Swirl Angle'
] + flowVelocity_4Paths + ['Average Flow Velocity'] + speedOfSound_4Paths + [
    'Average Speed of Sound'
] + signalStrength_4Paths + turbulence_4Paths + [
    'Meter Performance'
] + signalQuality_4Paths + gain_4Paths + transitTime_4Paths + ['Class']

headerMeterD = [
    'Profile Factor', 'Symmetry', 'Crossflow'
] + flowVelocity_4Paths + speedOfSound_4Paths + signalStrength_4Paths + signalQuality_4Paths + gain_4Paths + transitTime_4Paths + [
    'Class'
]

# add comma to all spaces
with open('dataset/Meter A', 'r+') as f:
    text = f.read()
    f.seek(0)
    f.truncate()
    f.write(text.replace('\t', ','))

with open('dataset/Meter B', 'r+') as f:
    text = f.read()
    f.seek(0)
    f.truncate()
    f.write(text.replace('\t', ','))

with open('dataset/Meter D', 'r+') as f:
    text = f.read()
    f.seek(0)
    f.truncate()
    f.write(text.replace('\t', ','))

# read text file and convert to csv
df_meterA = pd.read_csv('dataset/Meter A')
df_meterA.to_csv('dataset/Meter A.csv', index=None)

df_meterB = pd.read_csv('dataset/Meter B')
df_meterB.to_csv('dataset/Meter B.csv', index=None)

df_meterC = pd.read_csv('dataset/Meter D')
df_meterC.to_csv('dataset/Meter D.csv', index=None)

# add columns header
df_meterA = pd.read_csv('dataset/Meter A.csv', header=None)
df_meterA.columns = headerMeterA

df_meterB = pd.read_csv('dataset/Meter B.csv', header=None)
df_meterB.columns = headerMeterB

df_meterD = pd.read_csv('dataset/Meter D.csv', header=None)
df_meterD.columns = headerMeterD

df_meterA
df_meterB
df_meterD
```
